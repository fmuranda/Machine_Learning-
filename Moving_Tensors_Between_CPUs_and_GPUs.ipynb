```python
import torch
print(torch.__version__)
```

```python
if torch.cuda.is_available(): device = "cuda"
else: device = "cpu"
print(device)
```

```python
tens_a = torch.tensor([[10,11,12,13],[14,15,16,17]],device=device)
tens_b = torch.tensor([[18,19,20,21],[22,23,24,25]],device=device)
multi_tens =tens_a * tens_b
print(multi_tens)
```
