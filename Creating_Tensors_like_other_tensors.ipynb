```python
#@title Creating tensors like other tensors
```

```python
import torch
```

```python
#@title Create tensor of shape (2,5) filled with ones
starting_tensor = torch.zeros(2,5)
starting_tensor
```

```python
#@title Create a tensor with the same size filled with random numbers
torch.manual_seed(111)
random_tensor = torch.rand_like(starting_tensor)
random_tensor
```

```python
#@title Create a tensor with the same shape filled with ones
torch.ones_like(random_tensor)
```

```python
#@title Create a tensor with the same shape filled with provided value
torch.full_like(random_tensor, 7)
```
