```python
import torch
```

```python
#@title Creat a tensor of shape (2,5) filled with ones
starting_tensor = torch.zeros(2,5)
starting_tensor
```

```python
#@title Create a tensor of a shape (2,5) filled with ones
torch.manual_seed(111)
rand_tensor = torch.rand_like(starting_tensor)
rand_tensor
```

```python
#@title Create a tensot with the same shape filled with ones
torch.ones_like(rand_tensor)
```

```python
#@title Create a tensor with the same shape filled with provided value
torch.full_like(starting_tensor, 7)
```
